1. What is Flask Framework?
   

What are the advantages of Flask Framework? 
Explanation: Flask is a micro web framework written in Python. It is classified as a "micro-framework" because it does not require particular tools or libraries. It has no database abstraction layer, form validation, or any other components where pre-existing third-party libraries provide common functions. However, it supports extensions that can add application features as if they were implemented in Flask itself.

Advantages of Flask:

Lightweight: It has a small core and easy-to-extend philosophy.

Flexibility: It gives the developer control over how to structure the application.

Integrated Unit Testing: It has built-in support for unit testing.

Secure: It creates secure cookies for client-side sessions.

Jinja2 Templating: It uses Jinja2 for building dynamic web pages.

Scalable: It can handle high-traffic applications efficiently.

Q2.
Create a simple Flask application to display 'Hello World!!'.

```python
from flask import Flask


app = Flask(__name__)


@app.route("/")
def hello_world():
    return "Hello World!!"

if __name__ == '__main__':
    
    
    app.run(host="0.0.0.0")
```

Q3. What is App routing in Flask?
Why do we use app routes?

Explanation: App Routing is the technique used to map specific URLs (Uniform Resource Locators) to specific functions in your application. In Flask, we use the route() decorator to bind a function to a URL.

Why we use App Routes:

Navigation: They allow the application to handle different requests (e.g., Home page vs. Contact page) based on the URL visited by the user.

Organization: They help organize the application logic by separating code for different parts of the website.

Dynamic URLs: They allow for variable rules, enabling dynamic content loading (e.g., /aman/profile/829).

Q5. What function is used in Flask for URL Building? 
Write a Python code to demonstrate the working of the url_for() function.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Home Page'

@app.route('/welcome')
def welcome():
    return 'Welcome Page'


with app.test_request_context():
    print("The URL for the 'home' function is: ", url_for('home'))
    print("The URL for the 'welcome' function is: ", url_for('welcome'))
```
