Q1. What is an API? Give an example, where an API is used in real life.

Answer: An API (Application Programming Interface) is a set of rules and protocols that allows different software applications to communicate with each other. It acts as an intermediary, taking a request from one application, delivering it to a system, and returning the response to the application.

Real-life Example: Google Pay (GPay) Based on the whiteboard diagram provided, Google Pay acts as a perfect example of an API in action:

The Scenario: When you use GPay to send money, the GPay app does not have direct access to the bank's main database (for security reasons).

The API Interaction:

Request: GPay sends an API request to the Bank's server. As seen in the diagram, this could be requests like check_account(), check_balance(), or transfer_amount().

Processing: The Bank's API receives the request, authenticates it, and processes the transaction.

Response: The Bank sends a response back to GPay (e.g., "True" for a successful balance check or transaction complete), and GPay updates your screen.

Q2. Give advantages and disadvantages of using API.
Answer:

Advantages:

Integration: APIs allow different platforms (like GPay and a Bank) to interact seamlessly, enabling data flow between distinct systems.

Efficiency: Developers don't need to write code from scratch. They can use APIs (like Google Maps API or Email APIs) to add functionality instantly.

Automation: APIs allow computers to manage work rather than people, which significantly speeds up workflows (e.g., automatically sending a confirmation email after a purchase).

Disadvantages:

Security Risks: Since APIs are gateways to a server, they serve as a potential entry point for hackers if not properly secured.

Dependency: If a third-party API provider (e.g., a weather data service) crashes or changes their rules, your application might stop working.

Complexity: implementing and maintaining APIs requires specialized knowledge and can become complex when dealing with multiple external services.

Q3. Differentiate between API and Web API.

Answer: The distinction between an API and a Web API lies primarily in their scope and communication medium. An API (Application Programming Interface) is a broad term for any software intermediary that allows two applications to talk to each other, regardless of where they are located. This includes local interactions, such as a Python script importing a math library on your laptop without needing the internet. In contrast, a Web API is a specific type of API that is designed to be accessed over the web using the HTTP protocol. While all Web APIs are APIs, not all APIs are Web APIs. A Web API explicitly connects a client and a server across a network, such as the interaction between the Google Pay app on your phone and a remote banking server to process a transaction.

Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

Answer:

REST Architecture: REST (Representational State Transfer) is an architectural style for building web services. As outlined in the lecture slide, RESTful services are based on the following principles:

Client-Server: Separates the user interface (Client) from the data storage (Server).

Stateless: The server does not store any client context between requests; every request must contain all necessary information.

Cacheable: Responses can be stored (cached) to improve performance.

Uniform Interface: Uses standard HTTP verbs (GET, POST, PUT, DELETE).

Layered System: Allows the architecture to be composed of hierarchical layers (e.g., security layer, caching layer).

SOAP Architecture: SOAP (Simple Object Access Protocol) is a protocol for exchanging structured information. It relies heavily on XML for message format and is known for its strict standards and built-in security (WS-Security).

Shortcomings of SOAP:

Complexity: SOAP is much more complex to implement than REST.

Performance: It uses XML solely, which makes the message size large (heavyweight). Parsing XML is slower and requires more bandwidth compared to REST's JSON.

Rigidity: It is very strict with standards, making it less flexible for modern, rapid development.

Q5. Differentiate between REST and SOAP.

Answer: The difference between REST and SOAP is fundamentally about their structure and flexibility. REST (Representational State Transfer) is an architectural style that provides guidelines for communication, emphasizing a lightweight, stateless approach using standard HTTP verbs like GET and POST. It is highly flexible and typically uses JSON, which is easy to parse and bandwidth-efficient, making it the preferred choice for mobile applications like Google Pay. On the other hand, SOAP (Simple Object Access Protocol) is a strict protocol rather than a style. It imposes rigid standards and relies exclusively on XML for messaging, which makes the data transmission heavier and slower to process. While REST focuses on resources and simplicity, SOAP focuses on operations and high-security standards, making SOAP more complex to implement and maintain compared to the user-friendly nature of REST.
