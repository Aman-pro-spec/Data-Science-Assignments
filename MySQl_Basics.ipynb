Q1. What is a database?
Differentiate between SQL and NoSQL databases. 
A database is an organized collection of structured information or data, typically stored electronically in a computer system. It acts as a centralized repository that allows for easy access, management, and updating of data. When distinguishing between SQL and NoSQL, the primary difference lies in their structure and scalability. SQL databases are relational, meaning they use a table-based structure with rows and columns and require a predefined schema, making them ideal for complex queries and transactional data (e.g., MySQL). In contrast, NoSQL databases are non-relational and use dynamic schemas for unstructured data, storing information in formats like documents, key-values, or graphs. While SQL databases typically scale vertically by increasing server power, NoSQL databases scale horizontally by adding more servers (e.g., MongoDB).

Q2. DDL and its Commands
DDL, or Data Definition Language, refers to the set of SQL commands used to define and modify the structure of database objects like tables and schemas. The CREATE command is used to generate a new database or table structure from scratch. The DROP command is used when an entire table or database needs to be deleted, removing both the structure and the data. The ALTER command allows you to modify an existing table structure, such as adding or renaming a column, without deleting the table. Finally, TRUNCATE is used to remove all records from a table while keeping the table structure intact, effectively resetting it.

Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.
Data Manipulation Language (DML) consists of commands used to manage and manipulate the data stored within the database schema objects. The INSERT command is utilized to add new rows of data into a table. The UPDATE command modifies existing records within a table based on specified conditions (e.g., changing a student's name). The DELETE command allows for the removal of specific rows or records from a table without deleting the table itself.

Q4. What is DQL? Explain SELECT with an example.
Data Query Language (DQL) is the component of SQL used specifically to fetch and retrieve data from a database. The primary command used in DQL is SELECT, which allows users to query the database and retrieve specific data, which is then returned in a result set. For instance, SELECT * FROM Students; retrieves all data columns for every row in the 'Students' table.

Q5. Explain Primary Key and Foreign Key.
A Primary Key is a unique identifier for each record in a database table; it ensures that every row is distinct and cannot contain NULL values (e.g., a unique Student ID). A Foreign Key is a field in one table that links to the Primary Key of another table. It acts as a cross-reference between the two tables, establishing a relationship and ensuring referential integrity between the data (e.g., a 'Course_ID' in a Student table linking to a specific course in the Courses table).

Q.6 Explain the cursor() and execute() method.
To connect Python to MySQL, one typically uses a connector library to establish a session. In this process, the cursor() method is essential as it creates a cursor object that acts as a pointer or handle to the database context, allowing the Python script to interact with the database. The execute() method is then called on this cursor object to compile and run specific SQL commands (like queries or table creations) directly from the Python code.

Q7. Order of Execution in SQL
Although an SQL query is written starting with SELECT, the database engine executes the clauses in a specific logical order to optimize the retrieval process. The execution order begins with FROM (and JOINs) to identify the data source. It then proceeds to WHERE to filter rows based on conditions, followed by GROUP BY to aggregate data, and HAVING to filter those aggregated groups. Only after these steps does the engine process the SELECT clause to determine which columns to display. Finally, ORDER BY sorts the results, and LIMIT restricts the number of rows returned.
