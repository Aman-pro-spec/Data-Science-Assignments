Q1: Explain the following with an example: Artificial Intelligence, Machine Learning, Deep Learning.
Artificial Intelligence (AI):
Explanation: AI is the broad science of mimicking human abilities to solve problems and make decisions. It is the umbrella term for any computer program that does something smart.
Example: Smart assistants like Siri or Alexa which can understand spoken language and respond.
Machine Learning (ML):
Explanation: ML is a subset of AI. It involves training computers to learn from data without being explicitly programmed for specific rules. Instead of coding "if X happens, do Y," you feed the computer data, and it learns the patterns.
Example: Netflix's recommendation engine, which learns from your viewing history to suggest new movies.
Deep Learning (DL):
Explanation: DL is a specialized subset of ML inspired by the structure of the human brain. It uses multi-layered artificial neural networks to solve complex problems.
Example: Self-driving cars recognizing pedestrians or traffic signs, or FaceID on iPhones.

Q2: What is supervised learning? List some examples of supervised learning.
Definition:
Supervised learning is a type of machine learning where the model is trained on labeled data. This means the training dataset includes both the input data (features) and the correct output (labels). The model learns to map the input to the output.
Examples:
Spam Filtering: The model is trained on emails labeled as "Spam" or "Not Spam."
House Price Prediction: predicting the price of a house based on features like size, location, and number of bedrooms (where past prices are known).
Medical Diagnosis: Classifying a tumor as benign or malignant based on labeled medical images.

Q3: What is unsupervised learning? List some examples of unsupervised learning.
Definition:
Unsupervised learning involves training a model on data that is unlabeled. The system tries to learn the patterns and the structure from the data without any explicit guidance on what the output should be.
Examples:
Customer Segmentation: Grouping customers based on purchasing behavior without knowing the groups beforehand.
Market Basket Analysis: Identifying that people who buy bread also tend to buy butter (Association rules).
Anomaly Detection: Detecting credit card fraud by finding transactions that don't fit the standard pattern.

Q4: What is the difference between AI, ML, DL, and DS?
AI (Artificial Intelligence): The broad field of creating smart intelligent machines.
ML (Machine Learning): A subset of AI where machines learn from data.
DL (Deep Learning): A subset of ML that uses neural networks for complex patterns.
DS (Data Science): An interdisciplinary field that uses scientific methods, processes, algorithms, and systems to extract knowledge and insights from data. DS uses AI, ML, and DL as tools to analyze data, but it also includes statistics, data visualization, and data engineering.

Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?
The main differences lie in the type of data they use and their learning goals:
1. Supervised Learning
Data: It relies entirely on labeled data. This means the training set includes both the input data and the corresponding correct output (the "answers").
Goal: The goal is to train the model to predict outcomes or classify data accurately for new, unseen inputs.
Feedback Mechanism: The model gets direct feedback during training; it knows immediately if its prediction matches the labeled "answer."
2. Unsupervised Learning
Data: It uses unlabeled data. The training set contains only inputs without any corresponding target outputs.
Goal: The goal is to explore the data to find hidden structures, patterns, or customer segments (clusters) on its own.
Feedback Mechanism: There is no feedback loop. The model does not know if its outputs are "correct" in the traditional sense; it only looks for mathematical patterns.
3. Semi-Supervised Learning
Data: It uses a hybrid approach, combining a small amount of labeled data with a large amount of unlabeled data.
Goal: It bridges the gap between the two other methods. It is primarily used when labeling data is too expensive or time-consuming. The small labeled portion guides the learning, while the large unlabeled portion helps the model generalize better.

Q6: What is train, test and validation split? Explain the importance of each term.
When building a model, we split our dataset into three parts:
Training Set:
Definition: The largest portion of the data (usually 60-80%).
Importance: This is used to actually teach the model. The model learns the patterns from this data.
Validation Set:
Definition: A subset used during the training process (usually 10-20%).
Importance: It is used to tune the model's parameters (hyperparameters). It acts as a "practice exam" to see how the model is performing before the final test, preventing overfitting.
Test Set:
Definition: Data kept completely hidden from the model until the very end (usually 10-20%).
Importance: It provides an unbiased evaluation of the final model. It tells us how the model will perform in the real world on data it has never seen before.

Q7: How can unsupervised learning be used in anomaly detection?
Unsupervised learning is powerful for anomaly detection because it does not require a dataset of "known anomalies" (which are often rare and hard to find).
How it works:
The algorithm analyzes a dataset of "normal" behavior (e.g., standard credit card spending patterns).
It groups (clusters) this data to understand what "normal" looks like.
When a new data point comes in (e.g., a massive purchase in a foreign country), the algorithm calculates the distance of this point from the "normal" clusters.
If the data point is very far away or does not fit into any cluster, it is flagged as an anomaly.
